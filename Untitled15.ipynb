```python
%pip install scikit-image
!pip install matplotlib
import matplotlib.pyplot as plt
import skimage
import pandas as pd
import numpy as np
import os
import imageio.v2
from skimage.transform import resize
from skimage.io import imread
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
categories=['cats','dogs']
input=[]
output=[]
data='/content/drive/MyDrive/training_set/training_set'
for i in categories:
    path=os.path.join(data,i)
    for j in os.listdir(path):
        if j.endswith('.jpg'):
          img_path=os.path.join(path,j)
          img=imageio.v2.imread(img_path)
          img_resized= resize(img,(100,100,3))
          input.append(img_resized.flatten())
        output.append(categories.index(i))
input_arr=np.array(input)
output_arr=np.array(output)
```

```python
print(len(input_arr))
print(len(output_arr))
df=pd.DataFrame(input_arr)
df['Target']=output_arr[1:420]
df.shape
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=72,stratify=y)
```

```python
param_grid={'C':[0.1,1,10],'gamma':[0.1,1,10],'kernel':['linear']}
svc=svm.SVC(probability=True,kernel='linear')
model=GridSearchCV(svc,param_grid)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accurary=accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred,target_names=['cats','dogs']))
```
